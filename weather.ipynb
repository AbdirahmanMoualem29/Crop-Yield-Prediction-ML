```python
null_pct = weather.apply(pd.isnull).sum()/weather.shape[0]
```

```python
valid_columns = weather.columns[null_pct < 0.05]
```

```python
weather = weather[valid_columns].copy()
```

```python
weather.columns = weather.columns.str.lower()
```
